# Scraping

```python
#import necessary packages
import time
from selenium import webdriver
import pandas as pd
import matplotlib.pyplot as 
import re

#create an instance of browser
driver = webdriver.Chrome()

#creating a dictionary for storing the information after scraping
jobs={"roles":[],
     "companies":[],
     "locations":[],
     "experience":[],
     "skills":[]}
```

```python
#Navigate to the Medium popular page
driver.get("https://medium.com/topic/popular")
time.sleep(3)
```

```python
#to scroll down the page to the bottom to load all the items on the page
for i in range(10):
    time.sleep(3)
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
```

```python
#Extract all the
posts={"title":[],
       "time":[],
       "author":[]}   
for post in driver.find_elements_by_css_selector(".gq.gr.gk.n.gs.gt.gu.gv.gw.al"):
    title=post.find_element_by_css_selector(".ap.q.ee.cf.ef.cg.fl.gy.gz.as.av.fm.ek.el.au").text
    author=post.find_element_by_css_selector(".cf.b.cg.ch.ci.cj.as.av.en.ek.el.au.ap.q").text.lower().strip()
    dat=post.find_element_by_css_selector(".cf.b.cg.ch.ci.cj.r.ck.cl").text.lower().strip()
    posts["title"].append(title)
    posts["time"].append(dat)
    posts["author"].append(author)
```

```python
#coverting it to a Data frame
posts_df=pd.DataFrame(posts)
posts_df.head()
```

```python
posts_df.shape
```

```python
posts_df[["author","publication","null"]]=posts_df.author.str.split(" in ",expand=True)
posts_df[["date","duration"]]=posts_df.time.str.split("\n·\n",expand=True)
```

```python
posts_df.head()
```

```python
posts_df["duration"]=[int(re.sub("\D", "",dur)) for dur in posts_df.duration]
```

```python
posts_df.head()
```

```python
posts_df=posts_df.drop(["time","null"],axis=1)
```

## Scraping other elements- Claps and Tags
#### We need to scrape these elements seperately because these are only visible  inside the article.

```python
#finding all the article links on the Medium popular page.
urls=[]
elems = driver.find_elements_by_tag_name('a')
for elem in elems:
    href = elem.get_attribute('href')
    if href is not None:
        urls.append(href)
```

```python
article_urls=[url for url in urls if "topic_page" in url and "operation" not in url]
article_urls=[url for url in article_urls if len(url)>90]
final_urls=list(set(article_urls))
```

```python
# Extract the title,claps,tags for each article
article={"title":[],
        "clap":[],
         "tag":[]
        }
i=0
for url in final.urls:
    driver.get(url)
    time.sleep(3)
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(3)
    try:
        try:
            article["title"].append(driver.find_element_by_tag_name("h1").text)
        except:
            article["title"].append([f.text for f in driver.find_elements_by_tag_name("h2")][1])
    except:
        article["title"].append("None")
    try:
        try:
            article["clap"].append([i.text for i in driver.find_elements_by_tag_name("button") if i.text not in ['',"Follow","Get this newsletter",'Not you?','Get started']][0])
        except:
            article["clap"].append([i.text for i in driver.find_elements_by_tag_name("button") if "claps" in i.text])
    except:
        article["clap"].append("None")
    try:
        tag=[]
        for elem in driver.find_elements_by_tag_name('a'):
            href = elem.get_attribute('href')
            if any(re.findall(r'\/tag|tagged', href, re.IGNORECASE)):
                tag.append(href.split("/")[-1])
        article["tags"].append(tag)
    except:
        article["tags"].append("None")

    print(i)
    i=i+1
   
```

```python
article_df=pd.DataFrame(article)
article_df
```
